```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fig.suptitle('Construction Spending by Type',fontsize='x-large', fontname = 'Cambria')

ax.plot(range(len(df_con['Total Construction'])),df_con['Total Construction'])
ax.plot(range(len(df_con['Private Construction'])),df_con['Private Construction'])
ax.plot(range(len(df_con['Public Construction'])),df_con['Public Construction'])
ax.set_xticks(range(0,len(df_con['Total Construction']),12))
ax.xaxis.set_ticklabels(df_con['Month-Year'][::12])
ax.legend(loc = 'upper right', prop = {'family':'Average', 'size':'medium'})
ax.xaxis.set_tick_params(which = 'both', top = 'off', bottom = 'on', labelbottom = 'on') 
ax.yaxis.set_label_text('Construction Spending', fontsize = 12, fontname = 'Cambria') 
ax.xaxis.set_label_text('Year',fontname='Cambria',fontsize = 12)
ax.yaxis.set_tick_params(which = 'both', right = 'off', left = 'on', labelleft = 'on')   
fig.set_size_inches((10,7))
```

```python
x = df_con.Month[0:12]
y1 = df_con['Total Construction'][0:12]

x_labels = ['','J','F','M','A','M','J','J','A','S','O','N','D']

fig, (ax1) = plt.subplots() 

fig.suptitle('Total Construction: 2002',fontsize='x-large', fontname = 'Cambria')

ax1.plot(x,y1,label='Total Construction')                                  
                               
for ax in fig.axes:
    ax.spines['right'].set_visible(False)                                               
    ax.spines['top'].set_visible(False)                                                    
    ax.set_xlim(0,x.max()+1)                                                              
    ax.set_ylim(y1.min(),1.1*y1.max())                                                          

for ax in fig.axes:
    ax.xaxis.set_label_text('Month',fontname='Cambria', fontsize = 12)
    ax.xaxis.set_ticks(range(0,13))
    ax.xaxis.set_ticklabels(x_labels)
    ax.xaxis.set_tick_params(which = 'both', top = 'off', bottom = 'on', labelbottom = 'on')  

for ax in fig.axes:
    ax.yaxis.set_label_text('Construction Spending',fontsize = 12, fontname = 'Cambria')    
    ax.yaxis.set_tick_params(which = 'both', right = 'off', left = 'on', labelleft = 'on')   
    
fig.set_size_inches(7,5)           
fig.tight_layout()                            
```
